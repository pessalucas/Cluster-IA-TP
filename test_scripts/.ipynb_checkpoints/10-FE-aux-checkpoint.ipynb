```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# load dataset
dataset = pd.read_csv('data/dataset_v1.csv')

# rows and columns of the data
print(dataset.shape)

# visualise the dataset
dataset.head(2)
```

```python
pd.set_option('display.max_rows', dataset.shape[0]+1)
dataset.head(250)
```

```python
# 138 => 2010-01 to 215 => 2019-12
```

```python
dataset = dataset.iloc[138:208]
```

```python
dataset.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
dataset.shape
```

```python
dataset
```

```python
dataset["monto_cred_uva"].fillna( 0 , inplace = True)
```

```python
index = 139
value_ant = 0
load = 0

while index < 208:
    dataset["precio_prom_x"][index-1] = dataset["precio_prom_x"][index]
    dataset["precio_prom_x"][index+1] = dataset["precio_prom_x"][index]
    index = index + 3
```

```python
dataset["precio_prom_x"][207] = 2903
```

```python
145/3
```

```python
for value in dataset["precio_prom_x"]:
    if math.isnan(value) and load == 0:
        print(value)
    else:
        dataset["precio_prom_x"][index-1] = value
        dataset["precio_prom_x"][index+1] = value
        load = 1
    index = index + 1
```

```python
# Completo las tasas de interes faltantes con datos de otra bibliografia:
# https://repositorio.cepal.org/bitstream/handle/11362/36970/32/Argentina_es.pdf
# La tasa de interes en el 2013 era de 17.14%.
# La tasa de interes en el 2014 era de 24%.
# La tasa de interes en el 2015 era de 24.9%.
```

```python
dataset.loc[138:143,'tasas_interes_pm'] = 17.14
dataset.loc[144:155,'tasas_interes_pm'] = 24
dataset.loc[156:166,'tasas_interes_pm'] = 24.9
```

```python
dataset
```

```python
# Ajustado por dolar
dataset["alq_prom_dolar"] = dataset["precio_prom_y"] / dataset["tipo_cambio_implicito_en_adrs"]
```

```python
dataset
```

```python
dataset.drop('precio_prom_y', axis=1, inplace=True)
dataset.drop('tipo_cambio_implicito_en_adrs', axis=1, inplace=True)
```

```python
dataset.to_csv("data/dataset_v2.csv")
```
